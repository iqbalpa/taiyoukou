```python
temp = df[['area_name_romaji', 'name_romaji', 'lat', 'lon']]
temp = temp.drop_duplicates(subset=['area_name_romaji', 'name_romaji']).reset_index(drop=True)
temp
```

> Harvesine Distance to calculate distance between two points of lat-lon

> https://www.movable-type.co.uk/scripts/latlong.html

```python
def euclidian_dist(lat1, lon1, lat2, lon2):
    return ((lat1 - lat2)**2 + (lon1 - lon2)**2) ** 0.5

def harvesine_dist(lat1, lon1, lat2, lon2):
    from math import radians, sin, cos, sqrt, atan2

    R = 6371.0  # Earth radius in kilometers

    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)

    a = sin(dlat / 2)**2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
for plant in hatsuden:
    plant_lat = plant["latitude"]
    plant_lon = plant["longitude"]
    temp["dist"] = temp.apply(lambda row: harvesine_dist(plant_lat, plant_lon, row["lat"], row["lon"]), axis=1)
    nearest = temp.loc[temp["dist"].idxmin()]
    print(f"Plant: {plant['plant_name']}")
    print(f"Nearest Amedas Station: {nearest['name_romaji']} in {nearest['area_name_romaji']}")
    print(f"Distance: {nearest['dist']:.6f} degrees")
    print()
```

```python
import pandas as pd
import plotly.express as px

def plot_temperature_time_series(df: pd.DataFrame):
    fig = px.line(
        df,
        x='timestamp',
        y='temperature',
        color='region',  # Separates lines by region
        title='Temperature Over Time by Region',
        labels={
            'timestamp': 'Time',
            'temperature': 'Temperature ($^\circ$C)',
            'region': 'Region'
        }
    )

    fig.update_layout(
        xaxis_title='Time',
        yaxis_title='Temperature ($^\circ$C)',
        legend_title='Region',
        hovermode='x unified'
    )
    
    fig.show() 

plot_temperature_time_series(df)
```
